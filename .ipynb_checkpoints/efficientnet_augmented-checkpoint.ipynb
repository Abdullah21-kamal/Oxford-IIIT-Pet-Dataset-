```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image_dataset_from_directory
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# local path in my laptop
BASE_DIR = r"C:\Users\HP Zbook 17 G6\Desktop\RD\oxford-iiit-pet\images"

# subdirectories for train, validation, and test sets
TRAIN_DIR = os.path.join(BASE_DIR, "train")
VAL_DIR = os.path.join(BASE_DIR, "validation")
TEST_DIR = os.path.join(BASE_DIR, "test")

# image size and batch size
IMG_SIZE = 224  # Image size for model input
BATCH_SIZE = 32  # Batch size for training
NUM_CLASSES = 35  # Number of pet classes

# if dataset paths exist
assert os.path.exists(TRAIN_DIR), "Training directory not found!"
assert os.path.exists(VAL_DIR), "Validation directory not found!"
assert os.path.exists(TEST_DIR), "Test directory not found!"

print("exist")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# ImageDataGenerator with augmentation for training
train_datagen = ImageDataGenerator(
    rescale=1./255,  # Normalize pixel values
    rotation_range=30,  # Random rotation up to 30 degrees
    width_shift_range=0.2,  # Random horizontal shift
    height_shift_range=0.2,  # Random vertical shift
    shear_range=0.2,  # Shear transformation
    zoom_range=0.2,  # Random zoom
    horizontal_flip=True,  # Flip images horizontally
    fill_mode="nearest"  # Fill missing pixels after transformation
)

# validation and test data should not be augmented, only rescaled
val_test_datagen = ImageDataGenerator(rescale=1./255)

# Load images from directories
train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="categorical"  # Multi-class classification
)

val_generator = val_test_datagen.flow_from_directory(
    VAL_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=False  # No shuffling for validation
)

test_generator = val_test_datagen.flow_from_directory(
    TEST_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=False  # No shuffling for test data
)

print("done")
```

```python
import matplotlib.pyplot as plt
```

```python
# augmented images
augmented_images, _ = next(train_generator)  # Extract only images, ignore labels

fig, axes = plt.subplots(2, 5, figsize=(15, 6))
axes = axes.flatten()

for img, ax in zip(augmented_images[:10], axes):
    ax.imshow(img)
    ax.axis("off")  # Hide axes

plt.tight_layout()
plt.show()
```

```python
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
```

```python

# using EfficientNetB0 pre-trained on ImageNet (excluding the top layer)
base_model = EfficientNetB0(weights="imagenet", include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))

# Freeze the base model layers
base_model.trainable = False

# Add custom classification layers
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(256, activation="relu")(x)
x = Dropout(0.3)(x)  # Dropout for regularization
output_layer = Dense(NUM_CLASSES, activation="softmax")(x)  # Final output layer

# Create the final model
model = Model(inputs=base_model.input, outputs=output_layer)

# Compile the model
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
model.summary()
```

```python
for layer in base_model.layers[:20]: 
    print(layer.name, layer.trainable)
```

```python
# Unfreeze some layers in the base model
for layer in base_model.layers[-50:]:  # Unfreezing last 50 layers
    layer.trainable = True

print("Unfrozen last 50 layers for fine-tuning.")

# Recompile the model with a lower learning rate
from tensorflow.keras.optimizers import Adam

fine_tune_lr = 1e-5  # Lower learning rate for fine-tuning
model.compile(optimizer=Adam(learning_rate=fine_tune_lr),
              loss="categorical_crossentropy",
              metrics=["accuracy"])

print("Model recompiled with lower learning rate.")
```

```python
fine_tune_epochs = 15  

history_fine = model.fit(
    train_generator,  # Augmented training data
    validation_data=val_generator,  # Validation set
    epochs=fine_tune_epochs,
    verbose=1
)
```

### classification Report

```python
from sklearn.metrics import classification_report
import numpy as np
```

```python
true_labels = test_generator.classes

# Predict the classes
predictions = model.predict(test_generator)
predicted_labels = np.argmax(predictions, axis=1)
class_names = list(test_generator.class_indices.keys())
report = classification_report(true_labels, predicted_labels, target_names=class_names)

print(report)
```

### **Analysis of Classification Report Output**
#### **Observations:**
1. **Overall Accuracy is Extremely Low (0.03 or 3%)**  
   - This indicates that the model is performing just slightly better than random guessing 
   
2. **Most Classes Have 0 Precision, Recall, and F1-score**  
   - The model fails to correctly classify almost all images
   - Only the "Egyptian_Mau" class shows recall of 1.00 (meaning it classifies all samples as this class

3. **Weighted and Macro Averages Show Poor Performance**  
   - A macro average (which treats all classes equally) of 0.03 means that the model has learned very little
   - A weighted average (which considers class distribution) of 0.03 also confirms the poor learning capability
