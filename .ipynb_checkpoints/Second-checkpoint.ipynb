```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define image size
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

# Define data augmentation
datagen = ImageDataGenerator(
    rescale=1./255,         # Normalize pixel values (0-1)
    rotation_range=20,      # Random rotation
    horizontal_flip=True,   # Flip images horizontally
    brightness_range=[0.8, 1.2],  # Adjust brightness
    validation_split=0.1    # 10% validation split
)

# Dataset paths
dataset_path = r"C:\Users\HP Zbook 17 G6\Desktop\RD\oxford-iiit-pet\images"

# Load training images
train_generator = datagen.flow_from_directory(
    directory=dataset_path,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training'  # Use training split
)

# Load validation images
val_generator = datagen.flow_from_directory(
    directory=dataset_path,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'  # Use validation split
)
```

```python
import os
import shutil

# Define dataset path
dataset_path = r"C:\Users\HP Zbook 17 G6\Desktop\RD\oxford-iiit-pet\images"

# List all images
images = os.listdir(dataset_path)

# Iterate over images and move them to respective breed folders
for img in images:
    if "_" in img:  # Ensure it's a valid image name
        breed = "_".join(img.split("_")[:-1])  # Extract breed name
        breed_folder = os.path.join(dataset_path, breed)

        # Create folder if it doesn't exist
        os.makedirs(breed_folder, exist_ok=True)

        # Move image to respective folder
        shutil.move(os.path.join(dataset_path, img), os.path.join(breed_folder, img))

print(" Images have been organized into class folders.")
```

```python
train_generator = datagen.flow_from_directory(
    directory=dataset_path,  # Now contains breed subfolders
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training'
)

val_generator = datagen.flow_from_directory(
    directory=dataset_path,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'
)
```

```python
import matplotlib.pyplot as plt

# Get a batch of images
sample_images, sample_labels = next(train_generator)

# Plot sample images
fig, axes = plt.subplots(1, 5, figsize=(15, 5))
for i in range(5):
    axes[i].imshow(sample_images[i])
    axes[i].axis("off")
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model

# Define input shape
IMG_SHAPE = (224, 224, 3)

# Load pretrained MobileNetV2 model (exclude top layer)
base_model = MobileNetV2(input_shape=IMG_SHAPE, include_top=False, weights='imagenet')

# Freeze base model layers (transfer learning)
base_model.trainable = False

# Add custom classification head
x = base_model.output
x = GlobalAveragePooling2D()(x)  # Convert feature maps to single vector
x = Dropout(0.5)(x)  # Reduce overfitting
x = Dense(128, activation='relu')(x)  # Fully connected layer
x = Dropout(0.3)(x)
predictions = Dense(37, activation='softmax')(x)  # 37 classes for pet breeds

# Create final model
model = Model(inputs=base_model.input, outputs=predictions)

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Model summary
model.summary()
```

```python
# Train the model
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=10,  # Start with 10 epochs (can increase if needed)
    verbose=1
)
```

```python
import matplotlib.pyplot as plt
```

```python


# Plot accuracy curves
plt.figure(figsize=(10, 5))
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.title('Model Accuracy')
plt.show()

# Plot loss curves
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title('Model Loss')
plt.show()
```

```python
import numpy as np
import seaborn as sns
from sklearn.metrics import confusion_matrix
```

```python


# Get true labels and predictions
y_pred = model.predict(val_generator)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = val_generator.classes

# Compute confusion matrix
cm = confusion_matrix(y_true, y_pred_classes)

# Plot confusion matrix
plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=False, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_true, y_pred_classes))
```

 2. Increase Model Training & Use Class Weights
Since some classes might be underrepresented, use class weighting:

 1. Train with Class Weights (Balance Training)
To fix class imbalance issues, use class weights:

```python
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

# Get class distribution
y_classes = train_generator.classes
class_weights = compute_class_weight(class_weight="balanced", classes=np.unique(y_classes), y=y_classes)

# Convert to dictionary format
class_weights_dict = dict(enumerate(class_weights))

# Train model with class weights
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=15,  # Increase epochs
    class_weight=class_weights_dict,  # Apply class balancing
    verbose=1
)
```

```python
# Plot accuracy curves
plt.figure(figsize=(10, 5))
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.title('Model Accuracy')
plt.show()

# Plot loss curves
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title('Model Loss')
plt.show()
```

 Next Steps
Since training with class weights did not improve classification performance, we need a stronger fix.

 1. Immediate Action: Fine-Tune MobileNetV2
Fine-tuning will allow deeper layers to learn breed-specific features.

```python
# Unfreeze last 30 layers for fine-tuning
base_model.trainable = True
for layer in base_model.layers[:-30]:  # Keep first 30 layers frozen
    layer.trainable = False

# Compile with lower learning rate
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001), 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])

# Train again with fine-tuning
history_finetune = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=10,  # Fine-tuning epochs
    verbose=1
)
```

 Why This?

Allows the model to extract deeper breed-specific features.
Prevents overfitting by freezing early layers.
The lower learning rate prevents drastic changes.
